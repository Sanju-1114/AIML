```python
import pandas as pd
```

```python
#DataFrame

#dictionary of lists

info = {
    "name": ["Sanju", "Patel", "Ayush", "Vinay"],
    "age": [20, 21, 22, 23],
    "gpa": [9.5, 8.5, 7.5, 9.2]
}

df = pd.DataFrame(info)
print(df)
print(df.index)
print(df.columns)
```

```python
# list of lists

df = pd.DataFrame([["Sanju", 20], ["Adam", 23], ["Eve" , 24]], columns=["Name", "Age"])
print(df)
```

```python
# from Numpy arrays

import numpy as np

np_arr = np.array([[1, 2, 3], [4, 5, 6]])

df = pd.DataFrame(np_arr, columns=["A", "B", "C"])

print(df)
```
