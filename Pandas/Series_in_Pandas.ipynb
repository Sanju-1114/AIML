```python
import pandas as pd
```

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))

#index
print(s[0])
print(s[2])
```

```python
#Labels
s = pd.Series([19, 20, 31, 43, 25], index = ["Sanju", "Vinay", "Ayush", "Patel", "Bob"])
print(s)

print(s["Sanju"])
print(s.index)
```

```python
# Vectorized Operations
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s1 + s2)
```
