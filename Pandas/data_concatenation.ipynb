```python
import pandas as pd
```

```python
# Concatenation

df1 = pd.DataFrame({
    "id" : [1, 2, 3],
    "name" : ["Sanju", "Eve", "Adam"]
})

df2 = pd.DataFrame({
    "id" : [4, 5, 6],
    "name" : ["Bob", "Charlie", "Gilian"]
})

# row wise concatenation
pd.concat([df1, df2])  
pd.concat([df1, df2], ignore_index=True)

# row wise concatenation
pd.concat([df1, df2], axis=1)
```
