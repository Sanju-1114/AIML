```python
import pandas as pd

df = pd.read_csv("raw_data.csv")
```

```python
# Cleaning Data

#Handle Missing Values
df.isnull()

df.isnull().sum()

df.dropna()  # row drop
# df.dropna(axis = 1)  # column drop

df.fillna(0)
df.fillna(100)

cleaned_data = df.copy()
cleaned_data["age"] = cleaned_data["age"].fillna(cleaned_data["age"].mean())

# df.ffill()
df.bfill()
```

```python
# Handle Duplicates

df.duplicated()
df["country"].duplicated()
df[["country", "gender"]].duplicated()
df2 = df.copy()

df.drop_duplicates() 
# df.drop_duplicates(inplace=True)   #Changes in original data
```

```python
#Handle Data Types & date-time

# Data Types
df.dtypes
df2 = df.copy()

df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes


#date-time
date_str = pd.Series([pd.to_datetime("2025-12-31")])
type(date_str.dtypes)
```
