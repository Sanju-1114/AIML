```python
import matplotlib.pyplot as plt
```

```python
# Histogram

import numpy as np
np.random.seed(0)

scores = np.random.normal(70, 10, 1000) # normal distribution

# bins = [0, 30, 70, 90, 100]
plt.hist(scores, color="skyblue", edgecolor="black", bins=10)
# plt.hist(scores, color="skyblue", edgecolor="black", bins=bins)
plt.xlabel("Scores")
plt.ylabel("# of Students")
```
