```python
import matplotlib.pyplot as plt
```

```python
#Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

#Plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
#Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Interception", "Hugo", "Lincoin"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(years, oscar_revenue, label="Oscar Movies") 
# plt.plot(years, oscar_revenue, "o--r", label="Oscar Movies") # format string
# plt.plot(years, oscar_revenue, color="red", label="Oscar Movies") 
# plt.plot(years, oscar_revenue, color="red", marker="o", label="Oscar Movies") 
# plt.plot(years, oscar_revenue, color="red", marker="o", linestyle="--", label="Oscar Movies")
# plt.plot(years, oscar_revenue, color="red", marker="o", linestyle="--", linewidth=3, label="Oscar Movies")

# plt.plot(years, non_oscar_revenue, "<-c", label="Non-Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
# plt.legend(loc="lower left")

# plt.style.available
# plt.style.use("dark_background")
# plt.style.use("ggplot")

plt.grid()
plt.show()
```

```python
with plt.xkcd():
    plt.plot(years, oscar_revenue, label="Oscar Movies")
    plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")
    
    plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
    plt.xlabel("Years")
    plt.ylabel("Revenue (in $M)")
    
    plt.legend()
    plt.show()
```
