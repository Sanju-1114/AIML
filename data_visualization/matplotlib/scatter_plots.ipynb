```python
import matplotlib.pyplot as plt
```

```python
#Scatter plots

people = ["PersonA", "PersonB", "PersonC", "PersonD", "PersonE",
          "PersonF", "PersonG", "PersonH", "PersonI", "PersonJ"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# colors = ["red", "purple", "yellow", "blue", "pink", "red", "purple", "yellow", "blue", "pink"]
colors = ["green" if x < 135 else "red" for x in blood_pressure]

# plt.scatter(age, blood_pressure)
# plt.scatter(age, blood_pressure, color="red")
# plt.scatter(age, blood_pressure, color=colors)
# plt.scatter(age, blood_pressure, color=colors, s=blood_pressure)
# plt.scatter(age, blood_pressure, color=colors, s=blood_pressure, marker="<")
# plt.scatter(age, blood_pressure, color=colors, s=blood_pressure, marker="o", alpha=0.7)

# plt.scatter(age, blood_pressure, s=blood_pressure, cmap="plasma", c=blood_pressure)
plt.scatter(age, blood_pressure, s=blood_pressure, cmap="OrRd", c=blood_pressure)

plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    # plt.annotate(people[i], xy=(age[i], blood_pressure[i]))
    # plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i], blood_pressure[i]))
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i]+1, blood_pressure[i]+1))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(blood_pressure), max(blood_pressure)+5)
    
```
