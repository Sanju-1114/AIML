```python
import matplotlib.pyplot as plt
```

```python
legit_transaction = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.55, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transaction = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]


# Transaction amount distribution - Legit vs Fraud

plt.hist(fraud_transaction, color="red", bins=15, label="Fraud Transactions", edgecolor="black", alpha=0.5)
plt.hist(legit_transaction, color="green", bins=15, label="Legit Transactions", edgecolor="black", alpha=0.5)

plt.title("Transaction amount distribution - Legit vs Fraud")
plt.xlabel("Amount (in $)")
plt.ylabel("Freqency")

plt.legend()
```
